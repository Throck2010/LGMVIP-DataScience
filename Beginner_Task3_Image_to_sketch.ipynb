### opencv is a wide open source library for computer vision and image processing.Various image and videos processing tasks can be carried out in python with the help of opencv library

```python
import cv2 as c
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
i = c.imread("thanos.jpg")
```

```python
cv2_imshow(i)
```

##**Image in RBG**

```python
rbg_i = c.cvtColor(i, c.COLOR_BGR2RGB)
cv2_imshow(rbg_i)
```

## **RBG Format to Grayscale Format**

```python
gs_i = c.cvtColor(rbg_i, c.COLOR_BGR2GRAY)
cv2_imshow(gs_i)
```

## **Grayscale Format to Inverted Grayscale Format(Negative Image)**

```python
igs  = 255 - gs_i
cv2_imshow(igs)
```

## **Blurry Image**

```python
bi = c.GaussianBlur(igs, (21,21),0)
cv2_imshow(bi)
```

## **Inverted Blurry Image**

```python
ib = 255 - bi
cv2_imshow(ib)
```

## **Pencil Sketch**

```python
ps = c.divide(gs_i,ib, scale = 255)
cv2_imshow(ps)
```

```python
cv2_imshow(i)
print("\n")
print("Original Image")
print("\n\n")
cv2_imshow(ps)
print("\n\n")
print("Pencil Sketch")
```
